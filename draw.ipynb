```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#transformer
barWidth = 0.3
 
# Choose the height of the blue bars
bars1 = [0.38418, 0.39334, 0.08882,0.03807,0.15035]
 
# Choose the height of the cyan bars
bars2 = [0.30395, 0.32079, 0.07194,0.03421,0.12446]
 
# Choose the height of the error bars (bars1)
yer1 = [0.5, 0.4, 0.5]
 
# Choose the height of the error bars (bars2)
yer2 = [1, 0.7, 1]
 
# The x position of bars
r1 = np.arange(len(bars1))
r2 = [x + barWidth for x in r1]
 
# Create blue bars
plt.bar(r1, bars1, width = barWidth, color = 'skyblue', edgecolor = 'black',  capsize=7, label='With Noise Input')
 
# Create cyan bars
plt.bar(r2, bars2, width = barWidth, color = 'darksalmon', edgecolor = 'black',  capsize=7, label='Origin Input')
 
# general layout
plt.xticks([r + barWidth for r in range(len(bars1))], ['R-MAE', 'R-MSE', 'T-MAE','T-MSE',"Geometric Mean"])
plt.ylabel('Value')
plt.legend()
 
# Show graphic
plt.show()
```
